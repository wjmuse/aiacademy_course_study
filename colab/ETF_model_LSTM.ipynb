```
import numpy as np
import pandas as pd
from collections import OrderedDict
from time import time

from prepare_data import get_sliding_window
```

```
 # read in csv, change column names
etf = pd.read_csv('twetf/data/tetfp.csv')
col_dtypes = OrderedDict(code=str, date=str, name=str, open=float, high=float, low=float, close=float, volume=int)

etf.columns=col_dtypes.keys()

etf.head()
```

```
etf= etf[:1286] # select 台灣50 
etf.tail()
```

```
etf['target'] = etf.close.diff().fillna(0)
etf['target_signal'] = np.where(etf.close.diff().fillna(0) >=0,1,-1) 
etf
```

```
from sklearn import preprocessing

def normalize(df):
    newdf= df.copy()
    min_max_scaler = preprocessing.MinMaxScaler()

    newdf['open'] = min_max_scaler.fit_transform(df.open.values.reshape(-1,1))
    newdf['low'] = min_max_scaler.fit_transform(df.low.values.reshape(-1,1))
    newdf['high'] = min_max_scaler.fit_transform(df.high.values.reshape(-1,1))
    newdf['volume'] = min_max_scaler.fit_transform(df.volume.values.reshape(-1,1))
    newdf['close'] = min_max_scaler.fit_transform(df.close.values.reshape(-1,1))
    newdf['target'] = min_max_scaler.fit_transform(df.target.values.reshape(-1,1))
    newdf['target_signal'] = min_max_scaler.fit_transform(df.target_signal.values.reshape(-1,1))
    #newdf['target_signal'] = df.target_signal
    return newdf
new_etf = normalize(etf)
```

```
new_etf
```

```
new_etf.columns 
```

```
feature_cols = ['open', 'high', 'low', 'close', 'volume']
```

```
def prepare_data(feature_df, feature_cols, target_df, input_days=20,output_days=5):
    
    
    '''
    feature_df: the compelet feature set 
    feature_cols : subset of features,   if needs to select subset from the complete features set
    
    
    '''
    reduce_feature_df = feature_df[feature_cols].iloc[:len(feature_df)-output_days] 
    print ('len_X',reduce_feature_df.shape)
    print (reduce_feature_df)
    
    target_y  = target_df.iloc[input_days:]
    
    print ('len_y',target_y.shape)
    print (target_y)
    
    data_X =get_sliding_window(reduce_feature_df, n_days=input_days)
    data_y = get_sliding_window(target_y, n_days=output_days)
    
    data_dict = {'training_X': data_X,
                'training_y': data_y}
    
    return data_dict
    
```

```
data = prepare_data(new_etf, feature_cols, new_etf[['target','target_signal']])
```

```
data['training_X'].shape
```

```
data['training_y'].shape
```

```
from sklearn.model_selection import train_test_split 

X = data['training_X']
Y = data['training_y'][:,:,1]  # only use target signal 
Y = Y[:,:,np.newaxis] # expand one more axis to align with NN outoput 
print (X.shape)
print (Y.shape)


#K_binary = keras.utils.to_categorical(K_label,2)
X_train, X_test, Y_train,Y_test = train_test_split (X,Y, test_size=0.2,shuffle=True)

#Y_train =  Y_train[:,:,np.newaxis]
#Y_test =  Y_test[:,:,np.newaxis]
#Y_train = np.array(Y_train)
print (X_train.shape)
print (Y_train.shape)
print (Y_train[:10])
```

```
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.layers.recurrent import LSTM
import keras


from keras.optimizers import Adam,SGD

from keras.layers import Input, Embedding, LSTM, Dense
from keras.models import Model
from keras.optimizers import Adam 
from keras.losses import binary_crossentropy
from keras.layers import Dropout
from keras.utils import to_categorical
from keras.layers import Reshape

from keras.layers import Lambda
```

```
#def build_model(input_length, input_dim):
input_length, input_dim = 20,5

d = 0.2

main_input = Input(shape=(input_length,input_dim),  name='main_input')
lstm_1 = LSTM(64, return_sequences=True, name='lastm_1')(main_input)
lstm_1 = Dropout(d)(lstm_1)
lstm_2 = LSTM(128, return_sequences=True, name='lastm_2')(lstm_1)
lstm_2 = Dropout(d)(lstm_2)
#lstm_3 = LSTM(128, return_sequences=False, name='lastm_3')(main_input)


lstm_3 = LSTM(128, return_sequences=False, name='lastm_3')(lstm_2)
lstm_3 = Dropout(d)(lstm_3)
logit = Dense(128, activation='relu', name='logit')(lstm_3)
logit = Dropout(d)(logit)

logit = Dense(64, activation='relu', name='logit_2')(logit)
logit = Dropout(d)(logit)


output = Dense(5, activation='sigmoid', name='output')(logit)
output = Dropout(d)(output)

# Using Lambda function to get multiple output 
# Need reshape to allign with Shape of Train_Y

output_0 = Lambda(lambda x: x[:,0])(output)
output_0 = Reshape((-1,))(output_0) 
output_1 = Lambda(lambda x: x[:,1])(output)
output_1 = Reshape((-1,))(output_1)
output_2 = Lambda(lambda x: x[:,2])(output)
output_2 = Reshape((-1,))(output_2)
output_3 = Lambda(lambda x: x[:,3])(output)
output_3 = Reshape((-1,))(output_3)
output_4 = Lambda(lambda x: x[:,4])(output)
output_4 = Reshape((-1,))(output_4)


#loss_type = 'mse'
loss_type = 'binary_crossentropy'
combine_loss = [loss_type]*5


model = Model(inputs=main_input, outputs=[output_0,output_1,output_2, output_3, output_4])
adam_op = Adam(lr=0.001)
model.compile(optimizer=adam_op, loss=combine_loss, metrics=['accuracy'])

print (model.summary())
```

```
print (Y_train.shape)

Y = [Y_train[:,i,:] for i in range(5)]  # simplify expression 
model.fit (X_train, Y, batch_size=64, epochs=500, validation_split=0.2,verbose=2)
```

```
Y_ = [Y_test[:,i,:] for i in range(5)]

model.evaluate(X_test,Y_)
```
